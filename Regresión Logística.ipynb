# Regresión Logística

Supongamos que queremos resolver el siguiente problema de clasificación. Supongamos que tenemos muestras a partir de la cual queremos modelar si un niño cuya edad es x, si está o no escolarizado.
Si y=1, el niño está escolarizado. Si y=0 el niño no está escolarizado.

```python
# imports
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
x=np.array([[1],[2],[3],[4],[5],[6]])
y=np.array([0,0,0,1,1,1])
clf = LinearRegression()
clf.fit(x, y)
y_pred = clf.predict(x)
plt.plot(x, y_pred,  color='blue')
plt.scatter(x, y,  color='black')
plt.show()
```

En principio alcanzaría con establecer que si la salida de la predicción (h) es mayor que 0.5, la salida será 1, y si es menor que 0.5 la salida será cero. Supongamos que ahora tenemos una medición mas, que es un chico de 12 años que está yendo a 7mo grado. Agregamos la muestra a nuestro dataset y recalculamos.

```python
x=np.array([[1],[2],[3],[4],[5],[6],[12]])
y=np.array([0,0,0,1,1,1,1])
clf = LinearRegression()
clf.fit(x, y)
y_pred = clf.predict(x)
plt.plot(x, y_pred,  color='blue')
plt.scatter(x, y,  color='black')
plt.show()
```

Para nuestro caso podemos predecir que el chico estará escolarizado si h > 4.1 y que no estará escolarizado si h < 4.1. En principio no parece que la regresión lineal sea una buena herramienta para encarar nuestro problema de regularización.
Para estos casos se utiliza la regresión logística, la hipótesis a partir de la cual establecemos nuestro modelo.
Modificaremos nuestra hipótesis $h_\theta(x)=\theta^T x$ y la modificaremos ligeramente a un caso mas genérico: $h_\theta(x)=g(\theta^T x)$. En el caso de la regresión logística:

$$g(z)=\frac{1}{1+e^{-z}}$$

Intuitivamente podemos decir que $g(z)$ es la probabilidad de que y=1 cuando los parámetros de entrada son x. Es decir:

$$g(z)=P(y=1|x;\theta)$$

Dado que se debe cumplir $P(y=0|x;\theta)+P(y=1|x;\theta)=1$, entonces podemos escribir a la probabilidad complementaria como:

$$P(y=0|x;\theta)=1-P(y=1|x;\theta)$$

Por lo tanto podemos analizar la salida $h_\theta(x)$ y decir que si $h_\theta(x)\geq0.5$ entonces y_pred=1. O bien y_pred=0 si $h_\theta(x)<0.5$. Donde $\theta$ representan el modelo general y $x$ las entradas particulares sobre las cuales me baso para hacer la predicción.

Ver [notebook](./logistic-regression-why-sigmoid.ipynb) sobre sigmoidea.


```python
from sklearn.linear_model import LogisticRegression
def model(x):
    return 1 / (1 + np.exp(-x))
x=np.array([[1],[2],[3],[4],[5],[6],[12]])
y=np.array([0,0,0,1,1,1,1])
clf = LogisticRegression(C=1e2)
clf.fit(x, y)
y_pred = clf.predict(x)
loss = model(x * clf.coef_ + clf.intercept_).ravel()
plt.plot(x, y_pred,  color='blue')
plt.plot(x, loss,  color='red')
plt.scatter(x, y,  color='black')
plt.show()
```
